```python
import csv

with open(f'../data/AVM_day0_spectra.csv', 'r') as spectra_csv:
    reader = csv.reader(spectra_csv, delimiter='\t')
    colnames = next(reader)
    colnames[0] = 'ppm'
    
    reader = csv.DictReader(spectra_csv, fieldnames=colnames, delimiter='\t')
    next(reader)
    for i, row in enumerate(reader):
        if i > 65000:
            print(row)
            break
```
